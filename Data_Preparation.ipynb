# Merging Severity Mild, Moderate and None into one Column as 2,3,1

```python
#Replacing Severity Value Moderate 3 and Mild 2
df_cleaned['Severity_Moderate'] = df_cleaned['Severity_Moderate'].replace(1, 3)
df_cleaned['Severity_Mild'] = df_cleaned['Severity_Mild'].replace(1, 2)
```

```python
display(df_cleaned.head())
```

```python
# Creating a new column 'Severity' that combines the three columns
df_cleaned['Severity'] = df_cleaned[['Severity_Mild', 'Severity_Moderate', 'Severity_None']].max(axis=1)
df_cleaned.to_csv('/Modified_Asthma_Dataset.csv', index=False)
df_cleaned.head()

#modified_file_path = 'Modified_Asthma_Dataset.csv'
#data_frame.to_csv(modified_file_path, index=False)

# Download the modified CSV file
#files.download(modified_file_path)
```
