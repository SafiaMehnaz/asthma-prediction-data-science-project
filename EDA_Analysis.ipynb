```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files

df=pd.read_csv('/content/Final_dataset.csv')
```

```python
print(df.columns)
```

```python
df.columns.to_list()
```

```python
#Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', cbar_kws={'shrink': 0.8}, center=0)
plt.title('Correlation Heatmap', fontsize=16)
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
#Target Variable Distribution
colors = ['#4682B4', '#808080']
plt.figure(figsize=(6, 4))
df['Target'].value_counts().plot(kind='bar', color=colors)
plt.title('Distribution of Target Variable')
plt.xlabel('Target')
plt.ylabel('Count')
plt.xticks(rotation=360, ha='right')
plt.show()
```

```python
#Symptom Frequency Distribution
symptom_columns = list(df.columns[:9]) + [df.columns[16]]

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

df[symptom_columns].sum().plot(kind='bar', figsize=(10, 4), color=colors)
plt.title('Symptom Frequency Distribution')
plt.xlabel('Symptoms')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Age Category Distribution
age_columns = df.columns[9:14]

colors = ['#4c72b0', '#55a868', '#c44e52', '#8172b2', '#ccb974']

df[age_columns].sum().plot(kind='bar', figsize=(8, 4), color=colors)
plt.title('Age Group Distribution')
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.xticks(rotation=360, ha='right')
plt.show()
```

```python
#Gender Distribution
gender_columns = ['Gender_Female', 'Gender_Male']
df[gender_columns].sum().plot(kind='bar', figsize=(6, 4), color=['pink', 'lightblue'])

plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=360, ha='right')
plt.show()
```
